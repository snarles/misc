```python
def get_words(ff, cutoff = None):
    wds = np.loadtxt(folder+ff, dtype=str, delimiter=';')
    wds_new = []
    for w in wds:
        w2 = w.split('. ')[1]
        if w2[0] == ' ':
            w2 = w2[1:]
        if w2[-1] == '.':
            w2 = w2[:-1]
        wds_new.append(w2.capitalize())
    # remove duplicates
    wds_new = np.array(wds_new)
    tmp, idx = np.unique(wds_new, return_index = True)
    wds_new = wds_new[np.sort(idx)]
    if cutoff is not None:
        return wds_new[:cutoff]
    return wds_new

def add_words_to_dict(dd, ws, i, chg):
    for w in ws:
        if not w in dd.keys():
            dd[w] = np.array([0., 0., 0., 0., 0., 0.])
        dd[w][i] = dd[w][i] + chg
```
